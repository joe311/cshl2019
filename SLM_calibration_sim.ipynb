```python
import numpy as np
from numpy.fft import fft2, ifft2, fftshift, ifftshift

%matplotlib notebook 
from matplotlib import pyplot as plt, gridspec
```

```python
imsize = (512, 512)

xx, yy = np.meshgrid(np.arange(imsize[0]), np.arange(imsize[1]))
slm_amp = np.zeros(imsize)
slm_amp[(xx-128)**2+(yy-256)**2<30**2] = 1
slm_amp[(xx-512+128)**2+(yy-256)**2<30**2] = 1
slm_phase = np.zeros(imsize)

slm_phase[:, 256:] = np.pi
# slm_phase[:, 256:] = xx[:, 256:] 

gs = gridspec.GridSpec(1, 2)
plt.figure()
plt.subplot(gs[0])
plt.imshow(slm_amp)
plt.title('SLM amp')
plt.subplot(gs[1])
plt.imshow(slm_phase)
plt.title('SLM phase')
```

```python
slm_field = slm_amp*np.exp(-1j*slm_phase)
target_field = fftshift(fft2(ifftshift(slm_field)))

gs = gridspec.GridSpec(1, 2)
plt.figure()
plt.subplot(gs[0])
plt.imshow(np.abs(target_field))
plt.axvline(256, ls='--',c='r', alpha=.5)
plt.axhline(256, ls='--',c='r', alpha=.5)
plt.title('Target amp')
plt.subplot(gs[1])
plt.imshow(np.angle(target_field))
plt.title('Target phase')
```
